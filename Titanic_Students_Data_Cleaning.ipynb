```python
import pandas as pd

titanic_df = pd.read_csv("/content/sample_data/Titanic_Dataset.csv")
titanic_df
```

```python
titanic_df.isnull().sum()
```

```python
# Fill missing Age with median
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

# Encode Sex column
titanic_df['Sex'] = titanic_df['Sex'].map({'male': 0, 'female': 1})

# Final check
titanic_df.isnull().sum()
```

```python
students_df = pd.read_csv("/content/sample_data/Students_Performance_Dataset.csv")
students_df
```

```python
students_df.isnull().sum()
```

```python
# Fill MathScore with mean
students_df['MathScore'].fillna(students_df['MathScore'].mean(), inplace=True)

# Encode Gender
students_df['Gender'] = students_df['Gender'].map({'male': 0, 'female': 1})

# Final check
students_df.isnull().sum()
```

```python
titanic_df.to_csv("/content/sample_data/Titanic_Dataset.csv", index=False)
students_df.to_csv("/content/sample_data/Students_Performance_Dataset.csv", index=False)
```
